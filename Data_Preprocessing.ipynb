```python
import pandas as pd
import sys
read_file = pd.read_csv('/Users/ramanlaldomadia/Desktop/Capstone/all-the-news-2-1.csv', nrows=1000)
read_file.shape
print("----Dataset Information----")
read_file.info()
print("----Glimpse of Dataset----")
print("An integer takes", sys.getsizeof(read_file), "bytes")
read_file.head(3)
```

```python
read_file=read_file.drop(columns=['section','Unnamed: 0.1','Unnamed: 0', 'url'])
print(read_file.columns)
```

```python
read_file['date'] = read_file['date'].str[-8:]
read_file=read_file.rename(columns={"date": "time"})
read_file.head(5)
```

```python
read_file['month']=read_file['month'].astype(int)
read_file['title']=read_file['title'].astype(str)
read_file['author']=read_file['author'].astype(str)
read_file['article']=read_file['article'].astype(str)
read_file['publication']=read_file['publication'].astype(str)
read_file.dtypes
```

```python
#List of Publications
read_file['publication'].unique()
```

```python
#Combine Varients 

#Combine Vice and Vice News
```

```python
#List of Top Publications
a=15 #Top n number of publications in the dataset
s=read_file['publication'].value_counts().nlargest(a).index.tolist()
print("List of Top Publications:", s)
```

```python
#keep top 15 (of 27) publications, remove everything else
read_file=read_file.loc[read_file['publication'].isin(s)]
read_file.head(5)
```

```python
#Reorder Variables
read_file = read_file[["month", "day", "year", "time", "author", "title", "article", "publication"]]
read_file

#Create Date variable - month, day, year
#Clean up: Drop month, day, year
read_file['date'] = read_file[read_file.columns[:3]].apply(
    lambda x: '/'.join(x.dropna().astype(str)),
    axis=1)
read_file=read_file.drop(columns=['day','month','year'])
read_file = read_file[["date", "time", "author", "title", "article", "publication"]]

read_file

#Create Date/Time variable - month, day, year, time
#Clean up: Drop time
read_file['date/time'] = read_file[read_file.columns[:2]].apply(
    lambda x: ' '.join(x.dropna().astype(str)),
    axis=1
)
read_file=read_file.drop(columns=['time'])
read_file = read_file[["date", 'date/time', "author", "title", "article", "publication"]]
read_file
```

```python
read_file.nunique()
```

```python
#Drop Duplicates
read_file=read_file.drop_duplicates()
read_file
```

```python
#Drop no data ----- HELPPP!!! Ideally we want all observations to have all data.
read_file.fillna(0)
```
